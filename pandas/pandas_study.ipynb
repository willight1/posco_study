## ‚úÖ 1. Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Íµ¨Ï°∞ ÌôïÏù∏

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)
print(df)

# ÏÉÅÎã® Ï≤´Ï§ÑÎ∂ÄÌÑ∞ ÏàúÏ∞®Ï†ÅÏù∏ ÌùêÎ¶ÑÏûÖÎãàÎã§.

# 'pandas'ÎùºÎäî Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù ÎèÑÍµ¨Î•º Î∂àÎü¨Ïò¥. 'pd'Îäî Ï§ÑÏó¨ÏÑú Î∂ÄÎ•¥Îäî Ïù¥Î¶Ñ(alias)
# 'data'ÎùºÎäî Ïù¥Î¶ÑÏùò Î≥ÄÏàòÏóê ÎîïÏÖîÎÑàÎ¶¨ ÌòïÏãùÏúºÎ°ú Îç∞Ïù¥ÌÑ∞Î•º Ï†ÄÏû•Ìï®
# 'Name'Ïù¥ÎùºÎäî Ïó¥(column)Ïóê 3Î™ÖÏùò Ïù¥Î¶ÑÏùÑ Î¶¨Ïä§Ìä∏Î°ú Ï†ÄÏû•
# 'Age'ÎùºÎäî Ïó¥ÏóêÎäî Í∞Å ÏÇ¨ÎûåÏùò ÎÇòÏù¥Î•º Ïà´Ïûê(int)Î°ú Ï†ÄÏû•
# 'City'ÎùºÎäî Ïó¥ÏóêÎäî Í∞Å ÏÇ¨ÎûåÏùò ÎèÑÏãú Ïù¥Î¶ÑÏùÑ Ï†ÄÏû•
# pd.DataFrame(): pandasÏùò DataFrameÏù¥ÎùºÎäî Ìëú ÌòïÏãù(ÏóëÏÖÄÏ≤òÎüº ÏÉùÍ∏¥ Íµ¨Ï°∞)ÏúºÎ°ú Î≥ÄÌôòÌï®
# 'df'Îäî Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏù¥ÎùºÎäî ÎúªÏúºÎ°ú ÏûêÏ£º Ïì∞Ïù¥Îäî Ïù¥Î¶Ñ
# print(): ÌôîÎ©¥Ïóê Ï∂úÎ†•ÌïòÎùºÎäî Îúª. Ìëú ÌòïÌÉúÎ°ú Î≥¥Ïó¨Ï§å
```

---
## ‚úÖ 2. Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏóÖÎç∞Ïù¥Ìä∏ - ÎÇòÏù¥ ÎçîÌïòÍ∏∞, Ïó¥ Ï∂îÍ∞Ä, Ìñâ Ï∂îÍ∞Ä Îì±

```python
df = pd.DataFrame(data)

df['Age'] = df['Age'] + 1
df['Job'] = ['Engineer', 'Artist', 'Doctor']
df.loc[df['Name'] == 'Bob', 'City'] = 'San Francisco'
new_row = {'Name': 'David', 'Age': 28, 'City': 'Seattle', 'Job': 'Teacher'}
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
print(df)

# ÎîïÏÖîÎÑàÎ¶¨ 'data'Î•º Îã§Ïãú DataFrameÏúºÎ°ú ÎßåÎì¶
# 1. ÎÇòÏù¥Ïóê 1ÎçîÌïòÍ∏∞
# df['Age']: 'Age' Ïó¥ÏùÑ ÏÑ†ÌÉù / +1 : Í∞Å ÎÇòÏù¥Ïóê 1ÏùÑ ÎçîÌïúÎã§
# 2. ÏÉàÎ°úÏö¥ Ïó¥ Ï∂îÍ∞Ä : Job
# 'Job'Ïù¥ÎùºÎäî ÏÉà Ïó¥ÏùÑ ÎßåÎì§Í≥† Í∞Å ÏÇ¨ÎûåÏùò ÏßÅÏóÖÏùÑ ÏûÖÎ†•
# 3. BobÏùò ÎèÑÏãúÎ•º 'San Francisco'Î°ú Î∞îÍæ∏Í∏∞
# loc: Ìñâ ÏúÑÏπò ÏßÄÏ†ï / Ï°∞Í±¥: Ïù¥Î¶ÑÏù¥ 'Bob'Ïù¥Î©¥ / 'City' Í∞íÏùÑ Î≥ÄÍ≤Ω
# 4. ÏÉàÎ°úÏö¥ ÏÇ¨Îûå Ï∂îÍ∞Ä
# ÎîïÏÖîÎÑàÎ¶¨Î°ú ÏÉà ÏÇ¨Îûå Ï†ïÎ≥¥ ÏÉùÏÑ±
# pd.DataFrame([new_row]): Ìñâ ÌïòÎÇòÏßúÎ¶¨ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
# Í∏∞Ï°¥ dfÏôÄ ÏÉàÎ°úÏö¥ ÌñâÏùÑ Ìï©ÏπòÍ≥† Ïù∏Îç±Ïä§Î•º Îã§Ïãú ÏßÄÏ†ï
# Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ï∂úÎ†•
```

---
## ‚úÖ 3. 2Î≤àÏß∏ ÌñâÏóê ÏÉà Îç∞Ïù¥ÌÑ∞ ÏÇΩÏûÖ - ÏàúÏÑú ÏÑ§Î™Ö

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})
new_row = pd.DataFrame({'Name': ['David'], 'Age': [28]})
df_top = df.iloc[:2]
df_bottom = df.iloc[2:]
df = pd.concat([df_top, new_row, df_bottom], ignore_index=True)
print(df)

# 'New_row' ÏÉà Ìñâ ÎßåÎì§Í∏∞
# 'df_top' dfÏùò ÏúÑÏ™ΩÏóê 2Ìñâ Ï∂îÏ∂ú( 'iloc'ÎûÄ? : Ïà´ÏûêÎ°ú Ìñâ ÏÑ†ÌÉù (integer-locationÏùò Ï§ÑÏûÑÎßê))
# 'df_bottom' dfÏùò ÏïÑÎûòÏ™Ω 1Ìñâ Ï∂îÏ∂ú
# 'ÏúÑ' + 'ÏÉà Ìñâ' + 'ÏïÑÎûò Ìñâ' Ìï©ÏπòÍ∏∞, Ïù∏Îç±Ïä§ ÏÉàÎ°ú ÎßåÎì§Í∏∞( 'concat'Ïù¥ÎûÄ? : concatenate(Ïó∞Í≤∞)Ïùò Ï§ÑÏûÑÎßê. Ïó¨Îü¨ Í∞úÏùò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Î∂ôÏù¥Îäî Ìï®Ïàò))
# 'df' Ï∂úÎ†•
```

## ‚úÖ 4. Ïó¥(column) Ï∂îÍ∞ÄÌïòÍ∏∞ - Gender Ïó¥ Ï∂îÍ∞Ä

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})
df['Gender'] = ['Female', 'Male', 'Male']  # 'Gender'ÎùºÎäî Ïó¥ÏùÑ Ï∂îÍ∞ÄÌïòÍ≥† ÏÑ±Î≥Ñ Îç∞Ïù¥ÌÑ∞ ÏûÖÎ†•
print(df)
```

## ‚úÖ 5. Ìñâ(row) ÏÇ≠Ï†úÌïòÍ∏∞ - Bob ÏÇ≠Ï†ú

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})
df_row_deleted = df.drop(index=1)  # drop(): ÏÇ≠Ï†úÌï® / index=1: Îëê Î≤àÏß∏ Ìñâ (Bob)ÏùÑ ÏÇ≠Ï†ú
print("[Ìñâ ÏÇ≠Ï†ú] Ïù∏Îç±Ïä§ 1 (Bob) ÏÇ≠Ï†ú:")
print(df_row_deleted)
```

## ‚úÖ 6. Ïó¥(column) ÏÇ≠Ï†úÌïòÍ∏∞ - Age ÏÇ≠Ï†ú

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})
df_column_deleted = df.drop(columns=['Age'])  # drop(): ÏÇ≠Ï†ú / columns=['Age']: 'Age'ÎùºÎäî Ïó¥ÏùÑ ÏÇ≠Ï†úÌï®
print("[Ïó¥ ÏÇ≠Ï†ú] 'Age' Ïó¥ ÏÇ≠Ï†ú:")
print(df_column_deleted)
```

## ‚úÖ matplotlib
### ‚Ä¢ Í∏∞Î≥∏ ÏÇ¨Ïö©Î≤ï
‚Ä¢ plt.figure(figsize=(Í∞ÄÎ°ú, ÏÑ∏Î°ú)): Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ ÏÑ§Ï†ï

‚Ä¢ plt.title( ): Í∑∏ÎûòÌîÑ Ï†úÎ™©

‚Ä¢ plt.xlabel( ): xÏ∂ï Î†àÏù¥Î∏î

‚Ä¢ plt.ylabel( ): yÏ∂ï Î†àÏù¥Î∏î

‚Ä¢ plt.legend(): Î≤îÎ°Ä

‚Ä¢ plt.show(): Í∑∏ÎûòÌîÑÎ•º ÌôîÎ©¥Ïóê ÌëúÏãú

```python
import matplotlib.pyplot as plt

x = ['Ïõî', 'Ìôî', 'Ïàò', 'Î™©', 'Í∏à']
y = [100, 120, 90, 110, 130]

plt.figure(figsize=(8, 4))             # Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ ÏÑ§Ï†ï
plt.plot(x, y, label="Î∞©Î¨∏Ïûê Ïàò")       # ÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.title("ÏöîÏùºÎ≥Ñ Î∞©Î¨∏Ïûê Ïàò")          # Ï†úÎ™© Î∂ôÏù¥Í∏∞
plt.xlabel("ÏöîÏùº")                     # xÏ∂ï Ïù¥Î¶Ñ
plt.ylabel("Î™Ö Ïàò")                    # yÏ∂ï Ïù¥Î¶Ñ
plt.legend()                           # Î≤îÎ°Ä ÌëúÏãú
plt.show()                             # Í∑∏ÎûòÌîÑ Ï∂úÎ†•
```

```python
# üîß ÏãúÍ∞ÅÌôî ÎèÑÍµ¨ seabornÍ≥º matplotlib Î∂àÎü¨Ïò§Í∏∞
import seaborn as sns        # seabornÏùÄ pandas Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÍ≥º Ïûò Ïñ¥Ïö∏Î¶¨Îäî ÏãúÍ∞ÅÌôî ÎèÑÍµ¨ÏûÖÎãàÎã§
import matplotlib.pyplot as plt  # matplotlibÏùÄ Í∑∏ÎûòÌîÑÎ•º ÌôîÎ©¥Ïóê Ï∂úÎ†•Ìï† Îïå ÏÇ¨Ïö©Ìï©ÎãàÎã§

# üìä ÏòàÏãú Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
x = [1, 2, 3, 4]     # xÏ∂ï Í∞íÏûÖÎãàÎã§: 1, 2, 3, 4
y = [2, 4, 6, 8]     # yÏ∂ï Í∞íÏûÖÎãàÎã§: xÍ∞Ä Ï¶ùÍ∞ÄÌï†ÏàòÎ°ù yÎèÑ Ï¶ùÍ∞ÄÌïòÎäî ÏÑ†Ìòï Îç∞Ïù¥ÌÑ∞

# üß± Îç∞Ïù¥ÌÑ∞Î•º Ìëú ÌòïÌÉúÎ°ú ÎßåÎì§Í∏∞ (pandas DataFrame ÏÇ¨Ïö©)
import pandas as pd
df = pd.DataFrame({'x': x, 'y': y})  # 'x'ÏôÄ 'y'ÎùºÎäî Ïó¥(column)ÏùÑ Í∞ÄÏßÑ ÌëúÎ•º ÎßåÎì§Ïñ¥Ï§çÎãàÎã§

# üìà Ï≤´ Î≤àÏß∏ Í∑∏ÎûòÌîÑ: xÏ∂ïÍ≥º yÏ∂ïÏùÑ Î™ÖÌôïÌûà ÏßÄÏ†ïÌï¥ÏÑú ÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
sns.lineplot(data=df, x='x', y='y')  # seabornÏùò lineplotÏùÄ ÏÑ†Ìòï Í∑∏ÎûòÌîÑÎ•º Í∑∏Î†§Ï§çÎãàÎã§
plt.title("xÏôÄ yÎ•º ÏßÄÏ†ïÌïú ÏÑ† Í∑∏ÎûòÌîÑ")   # Í∑∏ÎûòÌîÑ Ï†úÎ™©
plt.show()  # Í∑∏ÎûòÌîÑÎ•º Ïã§Ï†úÎ°ú ÌôîÎ©¥Ïóê Î≥¥Ïó¨Ï§çÎãàÎã§

# üìà Îëê Î≤àÏß∏ Í∑∏ÎûòÌîÑ: Îç∞Ïù¥ÌÑ∞Îßå ÎÑòÍ∏∞Í≥† seabornÏù¥ ÏûêÎèôÏúºÎ°ú x, yÎ•º Ï∂îÏ†ïÌïòÍ≤å Ìï®
sns.lineplot(data=df)  # Ïó¥ ÏàúÏÑúÎåÄÎ°ú ÏûêÎèôÏúºÎ°ú x='x', y='y'Î°ú Ïù∏ÏãùÌï®
plt.title("Ïó¥ ÏàúÏÑú Í∏∞Î∞ò ÏûêÎèô ÏÑ† Í∑∏ÎûòÌîÑ")  # Í∑∏ÎûòÌîÑ Ï†úÎ™©
plt.show()  # Í∑∏ÎûòÌîÑ Ï∂úÎ†•
```

## üé® matplotlib vs seaborn

Îç∞Ïù¥ÌÑ∞Î•º ÏãúÍ∞ÅÌôîÌï† Îïå ÏÇ¨Ïö©ÌïòÎäî ÎåÄÌëúÏ†ÅÏù∏ ÎùºÏù¥Î∏åÎü¨Î¶¨ Îëê Í∞ÄÏßÄÏùò Ï∞®Ïù¥Î•º ÏïåÏïÑÎ¥ÖÎãàÎã§.

---

### üß∞ matplotlib

- **Îã§ÏñëÌïú Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ**(Ïà´Ïûê, Î¨∏Ïûê Îì±)ÏùÑ Ï°∞Ìï©Ìï¥ÏÑú Í∑∏Î¶¥ Ïàò ÏûàÏäµÎãàÎã§.
- Î≥¥ÌÜµ **xÏ∂ï, yÏ∂ïÏùÑ Í∏∞Î∞òÏúºÎ°ú Ìïú 2Ï∞®Ïõê Í∑∏ÎûòÌîÑ**Î•º ÎßéÏù¥ ÏÇ¨Ïö©Ìï©ÎãàÎã§.
- **ÎßâÎåÄ Í∑∏ÎûòÌîÑ**, **ÏÑ† Í∑∏ÎûòÌîÑ**, **ÏÇ∞Ï†êÎèÑ** Í∞ôÏùÄ Í∏∞Î≥∏ Í∑∏ÎûòÌîÑÏóê Í∞ïÌï©ÎãàÎã§.
- **Ïª§Ïä§ÌÑ∞ÎßàÏù¥Ïßï(ÏÇ¨Ïö©Ïûê ÏÑ§Ï†ï)** Í∏∞Îä•Ïù¥ Îß§Ïö∞ ÌíçÎ∂ÄÌï©ÎãàÎã§. (Í∏ÄÍº¥, ÏÉâÏÉÅ, ÎààÍ∏à, Î∞∞Í≤Ω Îì± ÎßàÏùåÎåÄÎ°ú Ï°∞Ï†à Í∞ÄÎä•)
- Î≥µÏû°Ìïú Í∑∏ÎûòÌîÑÎÇò ÎåÄÏö©Îüâ Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôîÏóê Ï†ÅÌï©Ìï©ÎãàÎã§.
- ÌïòÏßÄÎßå **ÏΩîÎìúÍ∞Ä Í∏∏Í≥†, Î©îÎ™®Î¶¨Î•º Îçî ÎßéÏù¥ ÏÇ¨Ïö©**ÌïòÎ©∞ ÏÜçÎèÑÎèÑ Ï°∞Í∏à ÎäêÎ¶¥ Ïàò ÏûàÏäµÎãàÎã§.

---

### üåà seaborn

- **pandasÏùò DataFrame**Ï≤òÎüº Í∞ôÏùÄ ÌòïÏãùÏùò Îç∞Ïù¥ÌÑ∞Î•º Í∏∞Î∞òÏúºÎ°ú ÏûëÎèôÌï©ÎãàÎã§.
- **ÌÜµÍ≥ÑÏ†ÅÏù∏ ÏãúÍ∞ÅÌôî**Ïóê ÌäπÌôîÎêòÏñ¥ ÏûàÏäµÎãàÎã§. Ïòà: ÌèâÍ∑†ÏÑ†, Î∂ÑÌè¨ ÌûàÏä§ÌÜ†Í∑∏Îû®, ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Îì±
- **Îã§Ï∞®Ïõê Îç∞Ïù¥ÌÑ∞**(Ïó¨Îü¨ Ïª¨Îüº, Í∑∏Î£πÎ≥Ñ) ÏãúÍ∞ÅÌôîÎ•º Í∞ÑÎã®ÌïòÍ≤å Ï≤òÎ¶¨Ìï† Ïàò ÏûàÏäµÎãàÎã§.
- Í∏∞Î≥∏Ï†ÅÏúºÎ°ú **ÏòàÏÅú Ïä§ÌÉÄÏùº**Ïù¥ Ï†ÅÏö©ÎêòÏñ¥ ÏûàÏñ¥ÏÑú, Íæ∏ÎØ∏ÏßÄ ÏïäÏïÑÎèÑ Î≥¥Í∏∞ Ï¢ãÏäµÎãàÎã§.
- **ÏΩîÎìúÍ∞Ä ÏßßÍ≥† Í∞ÑÎã®ÌïòÎ©∞**, ÏÜçÎèÑÎèÑ matplotlibÎ≥¥Îã§ Îπ†Î¶ÖÎãàÎã§.
- ÌïòÏßÄÎßå Î≥µÏû°Ìïú Ïª§Ïä§ÌÑ∞ÎßàÏù¥ÏßïÏùÄ matplotlibÎ≥¥Îã§ Ïñ¥Î†µÏäµÎãàÎã§.

---

### ‚úÖ ÏöîÏïΩ

- `matplotlib`ÏùÄ **Ï†ïÎ∞ÄÌïú Ï†úÏñ¥**Í∞Ä ÌïÑÏöîÌï† Îïå Ï¢ãÏäµÎãàÎã§.  
- `seaborn`ÏùÄ **Îπ†Î•¥Í≥† Í∞ÑÎã®ÌïòÍ≤å ÌÜµÍ≥Ñ Î∂ÑÏÑù ÏãúÍ∞ÅÌôî**Ìï† Îïå Ï¢ãÏäµÎãàÎã§.

(Ïòà: seabornÏúºÎ°ú Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Í≥†, matplotlibÏúºÎ°ú Ï†úÎ™©/Ìè∞Ìä∏/ÏÉâÏÉÅ ÏàòÏ†ï)
