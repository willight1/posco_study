# Pandas 실습문제

### 1. CSV 파일을 읽고 상위 5개 데이터를 출력하시오.

```python
import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv("posco_study_data.csv")

# 상위 5개 행 출력
df.head()
```

### 2. 데이터의 행과 열의 개수, 컬럼명, 데이터 타입을 확인하시오.

```python
# 행과 열 개수
print("Shape:", df.shape)

# 컬럼명
print("Columns:", df.columns.tolist())

# 데이터 타입
print("Dtypes:\n", df.dtypes)
```

### 3. 데이터의 결측치(null) 개수를 확인하시오.

```python
# 결측치 개수 확인
df.isnull().sum()
```

### 4. 특정 컬럼(Age)의 평균, 중앙값, 최댓값, 최솟값을 구하시오.

```python
# Age 컬럼 통계
print("평균:", df['Age'].mean())
print("중앙값:", df['Age'].median())
print("최댓값:", df['Age'].max())
print("최솟값:", df['Age'].min())
```

### 5. Salary 컬럼 기준 내림차순 정렬을 수행하시오.

```python
# Salary 내림차순 정렬
df.sort_values(by='Salary', ascending=False)
```

### 6. Department가 'Sales'인 행만 추출하시오.

```python
# Department가 'Sales'인 행 필터링
df[df['Department'] == 'Sales']
```

### 7. Salary 컬럼의 결측값을 평균 값으로 채우시오. 결과는 상위 5개만 확인하시오.

```python
# Salary 평균 계산
mean_salary = df['Salary'].mean()

# 결측값 채우기
df['Salary'] = df['Salary'].fillna(mean_salary)

df.head()  # 결과 확인
```

### 8. Department별 평균 Salary와 평균 Age를 구하시오.

```python
# 부서별 평균 Salary와 평균 Age 계산
df.groupby('Department')[['Salary', 'Age']].mean()
```

### 9. Age 컬럼을 20대, 30대, 40대 그룹으로 나누어 Age_Group 컬럼을 만들고 평균 Salary를 구하시오.

```python
# Age 구간 나누기
bins = [19, 29, 39, 49]  # 20대, 30대, 40대
labels = ['20대', '30대', '40대']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Age_Group별 평균 Salary
df.groupby('Age_Group')['Salary'].mean()
```

### 10. 특정 조건 필터링
Department가 'Engineering'이면서 Salary가 5000 이상인 직원들의 이름과 나이를 출력하시오.

```python
# 조건 필터링: 부서가 Engineering이고 Salary >= 5000
filtered = df[(df['Department'] == 'Engineering') & (df['Salary'] >= 5000)]

# 이름과 나이 출력
print(filtered[['Name', 'Age']])
```

### 11. 나이 구간별 인원수 집계
Age 컬럼을 20대, 30대, 40대, 50대 그룹으로 나누어(pd.cut 사용) 각 구간별 인원수를 구하시오.

```python
# 구간 정의 및 라벨링
bins = [19, 29, 39, 49, 59]
labels = ['20대', '30대', '40대', '50대']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# 각 그룹별 인원수 계산
df['Age_Group'].value_counts()
```

### 12. 결측치 비율 확인
각 컬럼별 결측치 비율(%)을 계산하여 출력하시오.

```python
# 결측치 비율 계산
(df.isnull().sum() / len(df) * 100).round(2)
```

### 13. Salary 상위 10% 직원 추출
Salary 기준 상위 10% 직원의 Name, Department, Salary만 출력하시오.

```python
# 상위 10% 기준 Salary 값
threshold = df['Salary'].quantile(0.9)

# 상위 10% 데이터 필터링
df[df['Salary'] >= threshold][['Name', 'Department', 'Salary']]
```

### 14. 부서별 최고 연봉자 찾기
Department별 최고 연봉(Salary)을 받는 직원의 이름과 연봉을 출력하시오.

```python
# 부서별 최고 연봉자 구하기
idx = df.groupby('Department')['Salary'].idxmax()

# 최고 연봉자 행 출력
df.loc[idx, ['Department', 'Name', 'Salary']]
```

### 15. 중복 데이터 처리
Name이 중복되는 경우를 찾아 중복된 데이터만 출력하시오.

중복 제거 후의 데이터프레임도 출력하시오.

```python
# 중복된 Name 데이터 확인
duplicates = df[df.duplicated('Name', keep=False)]
print("중복된 데이터:\n", duplicates)

# 중복 제거 후 데이터프레임
df_no_duplicates = df.drop_duplicates('Name')
print("중복 제거 후 데이터프레임:\n", df_no_duplicates)
```

### 16. 다중 조건 정렬
Department를 오름차순으로, 같은 부서 내에서는 Salary를 내림차순으로 정렬하시오.

```python
df.sort_values(by=['Department', 'Salary'], ascending=[True, False])
```
