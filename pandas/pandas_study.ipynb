# 🐼 판다스(Pandas)란?
- Pandas는 Python에서 표(테이블) 형태의 데이터를 쉽게 다룰 수 있게 해주는 도구

# ✔ 이름 뜻?
- Pandas는 “Panel Data”에서 따온 말

- 원래 경제학·통계학에서 시간 + 여러 변수로 구성된 데이터를 다룰 때 쓰던 용어

# 📊 쉽게 말하면?
- 판다스는 엑셀처럼 생긴 표를 파이썬으로 다룰 수 있게 해주는 도구

---
## ✅ 1. 데이터프레임 구조 확인

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)
print(df)

# 상단 첫줄부터 순차적인 흐름입니다.

# 'pandas'라는 데이터 분석 도구를 불러옴. 'pd'는 줄여서 부르는 이름(alias)
# 'data'라는 이름의 변수에 딕셔너리 형식으로 데이터를 저장함
# 'Name'이라는 열(column)에 3명의 이름을 리스트로 저장
# 'Age'라는 열에는 각 사람의 나이를 숫자(int)로 저장
# 'City'라는 열에는 각 사람의 도시 이름을 저장
# pd.DataFrame(): pandas의 DataFrame이라는 표 형식(엑셀처럼 생긴 구조)으로 변환함
# 'df'는 데이터프레임이라는 뜻으로 자주 쓰이는 이름
# print(): 화면에 출력하라는 뜻. 표 형태로 보여줌
```

---
## ✅ 2. 데이터프레임 업데이트 - 나이 더하기, 열 추가, 행 추가 등

```python
df = pd.DataFrame(data)

df['Age'] = df['Age'] + 1
df['Job'] = ['Engineer', 'Artist', 'Doctor']
df.loc[df['Name'] == 'Bob', 'City'] = 'San Francisco'
new_row = {'Name': 'David', 'Age': 28, 'City': 'Seattle', 'Job': 'Teacher'}
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
print(df)

# 딕셔너리 'data'를 다시 DataFrame으로 만듦
# 1. 나이에 1더하기
# df['Age']: 'Age' 열을 선택 / +1 : 각 나이에 1을 더한다
# 2. 새로운 열 추가 : Job
# 'Job'이라는 새 열을 만들고 각 사람의 직업을 입력
# 3. Bob의 도시를 'San Francisco'로 바꾸기
# loc: 행 위치 지정 / 조건: 이름이 'Bob'이면 / 'City' 값을 변경
# 4. 새로운 사람 추가
# 딕셔너리로 새 사람 정보 생성
# pd.DataFrame([new_row]): 행 하나짜리 데이터프레임 생성
# 기존 df와 새로운 행을 합치고 인덱스를 다시 지정
# 전체 데이터프레임 출력
```

---
## ✅ 3. 2번째 행에 새 데이터 삽입 - 순서 설명

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})
new_row = pd.DataFrame({'Name': ['David'], 'Age': [28]})
df_top = df.iloc[:2]
df_bottom = df.iloc[2:]
df = pd.concat([df_top, new_row, df_bottom], ignore_index=True)
print(df)

# 'New_row' 새 행 만들기
# 'df_top' df의 위쪽에 2행 추출( 'iloc'란? : 숫자로 행 선택 (integer-location의 줄임말))
# 'df_bottom' df의 아래쪽 1행 추출
# '위' + '새 행' + '아래 행' 합치기, 인덱스 새로 만들기( 'concat'이란? : concatenate(연결)의 줄임말. 여러 개의 데이터프레임을 붙이는 함수))
# 'df' 출력
```

---
## ✅ 4. 열(column) 추가하기 - Gender 열 추가

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})
df['Gender'] = ['Female', 'Male', 'Male']  # 'Gender'라는 열을 추가하고 성별 데이터 입력
print(df)
```

---
## ✅ 5. 행(row) 삭제하기 - Bob 삭제

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})
df_row_deleted = df.drop(index=1)  # drop(): 삭제함 / index=1: 두 번째 행 (Bob)을 삭제
print("[행 삭제] 인덱스 1 (Bob) 삭제:")
print(df_row_deleted)
```

---
## ✅ 6. 열(column) 삭제하기 - Age 삭제

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})
df_column_deleted = df.drop(columns=['Age'])  # drop(): 삭제 / columns=['Age']: 'Age'라는 열을 삭제함
print("[열 삭제] 'Age' 열 삭제:")
print(df_column_deleted)
```

## ✅ matplotlib
### • 기본 사용법
• plt.figure(figsize=(가로, 세로)): 그래프 크기 설정

• plt.title( ): 그래프 제목

• plt.xlabel( ): x축 레이블

• plt.ylabel( ): y축 레이블

• plt.legend(): 범례

• plt.show(): 그래프를 화면에 표시

```python
import matplotlib.pyplot as plt

x = ['월', '화', '수', '목', '금']
y = [100, 120, 90, 110, 130]

plt.figure(figsize=(8, 4))             # 그래프 크기 설정
plt.plot(x, y, label="방문자 수")       # 선 그래프 그리기
plt.title("요일별 방문자 수")          # 제목 붙이기
plt.xlabel("요일")                     # x축 이름
plt.ylabel("명 수")                    # y축 이름
plt.legend()                           # 범례 표시
plt.show()                             # 그래프 출력
```

```python
# 🔧 시각화 도구 seaborn과 matplotlib 불러오기
import seaborn as sns        # seaborn은 pandas 데이터프레임과 잘 어울리는 시각화 도구입니다
import matplotlib.pyplot as plt  # matplotlib은 그래프를 화면에 출력할 때 사용합니다

# 📊 예시 데이터 준비
x = [1, 2, 3, 4]     # x축 값입니다: 1, 2, 3, 4
y = [2, 4, 6, 8]     # y축 값입니다: x가 증가할수록 y도 증가하는 선형 데이터

# 🧱 데이터를 표 형태로 만들기 (pandas DataFrame 사용)
import pandas as pd
df = pd.DataFrame({'x': x, 'y': y})  # 'x'와 'y'라는 열(column)을 가진 표를 만들어줍니다

# 📈 첫 번째 그래프: x축과 y축을 명확히 지정해서 선 그래프 그리기
sns.lineplot(data=df, x='x', y='y')  # seaborn의 lineplot은 선형 그래프를 그려줍니다
plt.title("x와 y를 지정한 선 그래프")   # 그래프 제목
plt.show()  # 그래프를 실제로 화면에 보여줍니다

# 📈 두 번째 그래프: 데이터만 넘기고 seaborn이 자동으로 x, y를 추정하게 함
sns.lineplot(data=df)  # 열 순서대로 자동으로 x='x', y='y'로 인식함
plt.title("열 순서 기반 자동 선 그래프")  # 그래프 제목
plt.show()  # 그래프 출력
```

## 🎨 matplotlib vs seaborn

데이터를 시각화할 때 사용하는 대표적인 라이브러리 두 가지의 차이를 알아봅니다.

---

### 🧰 matplotlib

- **다양한 데이터 타입**(숫자, 문자 등)을 조합해서 그릴 수 있습니다.
- 보통 **x축, y축을 기반으로 한 2차원 그래프**를 많이 사용합니다.
- **막대 그래프**, **선 그래프**, **산점도** 같은 기본 그래프에 강합니다.
- **커스터마이징(사용자 설정)** 기능이 매우 풍부합니다. (글꼴, 색상, 눈금, 배경 등 마음대로 조절 가능)
- 복잡한 그래프나 대용량 데이터 시각화에 적합합니다.
- 하지만 **코드가 길고, 메모리를 더 많이 사용**하며 속도도 조금 느릴 수 있습니다.

---

### 🌈 seaborn

- **pandas의 DataFrame**처럼 같은 형식의 데이터를 기반으로 작동합니다.
- **통계적인 시각화**에 특화되어 있습니다. 예: 평균선, 분포 히스토그램, 상관관계 등
- **다차원 데이터**(여러 컬럼, 그룹별) 시각화를 간단하게 처리할 수 있습니다.
- 기본적으로 **예쁜 스타일**이 적용되어 있어서, 꾸미지 않아도 보기 좋습니다.
- **코드가 짧고 간단하며**, 속도도 matplotlib보다 빠릅니다.
- 하지만 복잡한 커스터마이징은 matplotlib보다 어렵습니다.

---

### ✅ 요약

- `matplotlib`은 **정밀한 제어**가 필요할 때 좋습니다.  
- `seaborn`은 **빠르고 간단하게 통계 분석 시각화**할 때 좋습니다.

→ 실제로는 두 라이브러리를 **함께 사용**하는 경우가 많습니다!  
(예: seaborn으로 그래프를 그리고, matplotlib으로 제목/폰트/색상 수정)
